Plot for across same donor's regions

```r
ggplot(Hiatt, aes(x = region, y = T.C, group = donor, color = donor)) +
  geom_line(size = 1) +           # Connect points for the same donor
  geom_point(size = 3) +          # Show points
  theme_minimal() +
  labs(
    x = "Region",
    y = "Unique SNVs",
    color = "Donor",
    title = "Unique T.C Across Regions by Donor"
  ) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) + facet_wrap(~cohort)
```
