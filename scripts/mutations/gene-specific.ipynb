```python
all_recurrent_genes = []
recurrent = '/uufs/chpc.utah.edu/common/HIPAA/u1264408/u1264408/Git/SEMIColon/data/output/CellCut/results/recurrent.vcf'
vcf = VCF(recurrent)
for variant in vcf:
        CSQ = variant.INFO.get("CSQ")
        recurrence = variant.INFO.get("RECURRENCE")
        files = variant.INFO.get("FILES")
        files = files.replace("/uufs/chpc.utah.edu/common/HIPAA/u1264408/u1264408/Git/SEMIColon/data/output/CellCut/results/", "")
        files = files.replace(".annotated.vcf.gz", "")
        files = files.replace(",", "+")
        split_CSQ = CSQ.split("|")
        chrom = variant.CHROM
        pos = variant.POS
        ref = variant.REF
        alt = variant.ALT
        gene = split_CSQ[3]
        consequence = split_CSQ[1]
        impact = split_CSQ[2]
        if consequence != "intergenic_variant" and gene != "" and impact in ["HIGH"]:
            print(chrom, pos, ref, alt, gene, consequence, impact, recurrence, files)
            all_recurrent_genes.append(gene)
            #print(split_CSQ)
all_recurrent_genes = set(all_recurrent_genes)
```
