```python
import os
import csv

input_dir = "/uufs/chpc.utah.edu/common/HIPAA/u1264408/u1264408/Git/SEMIColon/data/output/CellCut/results/vaf_spectra_noVAF/txtfiles"
report_file = "/uufs/chpc.utah.edu/common/HIPAA/u1264408/u1264408/Git/SEMIColon/data/output/CellCut/results/vaf_spectra_VAFinterval/txtfiles/CtoAvariants_report.tsv"

with open(report_file, "w") as out:
    out.write("file_name\ttotal\tC>A_count\tC>A_lowVAF_count\tfraction_lowVAF\n")

    for fname in os.listdir(input_dir):
        if not fname.endswith("snv_count.txt"):  # adjust suffix if needed
            continue

        path = os.path.join(input_dir, fname)
        with open(path) as f:
            lines = f.read().strip().splitlines()

        # last line is the phrase
        last_line = lines[-1].strip()

        # skip first 2 lines (comment + header), skip last line (phrase)
        data_lines = lines[2:-1]

        CtoA_count = 0
        CtoA_low = 0

        for line in data_lines:
            cols = line.split("\t")
            if len(cols) < 5:
                continue
            chrom, pos, ref, alt, vaf = cols
            # print(chrom, pos, ref, alt, vaf)
            if ref == "C" and alt == "A":
                CtoA_count += 1
                if float(vaf) < 0.2:
                    CtoA_low += 1

        fraction = CtoA_low / CtoA_count if CtoA_count > 0 else "NA"

        out.write(f"{fname}\t{last_line}\t{CtoA_count}\t{CtoA_low}\t{fraction}\n")
```
